#Simple data frames

```python
sns.set(style='whitregrid')
```

```python
data = {'Passed': np.random.randint(0, 2, size=6),
        'Math': [85, 90, 78, 92, 88, 75],
        'Science': [92, 88, 85, 95, 90, 80],
        'English': [78, 85, 80, 88, 82, 70],
        'History': [88, 82, 90, 85, 92, 78]}

df_easy= pd.DataFrame(data)
display(df_easy)
```

# Adding new feature/column into data frame with avg value

```python
df_easy['Average'] = df_easy[['Math','Science','English','History']].mean(axis=1)
```

```python
print(df_easy)
```

```python
display(df_easy)
```

#Visualize Average marks Vs Passed

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Passed', y='Average', data=df_easy,palette='Set2')
plt.title('Average Marks by Pass/Fail')
plt.xlabel('Passed(0 = No, 1 = Yes)')
plt.ylabel('Average Marks')
plt.tight_layout()
plt.show()
#
```
